## Setting Up an Analysis Pipeline (Task 2)

#### Step 1: Load Data
The data must be loaded in in order to work on it

```python
data1 = pds.read_csv('../data/raw/Maths_formatted.csv', engine='python', on_bad_lines='skip')
data2 = pds.read_csv('../data/raw/Portuguese_formatted.csv', engine='python', on_bad_lines='skip')
display(data1)
display(data2)
```

#### Step 2: Combine Datasets
Next, the datasets are combined for ease of use

```python
data1["subject"] = "M"
data2["subject"] = "P"
rawData = pds.concat([data1,data2], ignore_index=True)
```

#### Step 3: Optimize Columns
Unnecessary columns must be removed to make the data more visually understandable, ambiguous columns should be renamed and column values must be updated to become more readable

```python
col_remove = ["address","famsize","Pstatus","Mjob","Fjob","guardian","traveltime","failures","schoolsup","activities","nursery","internet","famrel","romantic","freetime","goout","Dalc","Walc","health","absences"]
cleanData = rawData.drop(col_remove, axis=1)
value_names = {'sex':{'M':'Male', 'F':'Female'},'Medu':{0:'None', 1:'Primary', 2:'Grade 5-9', 3:'Secondary', 4:'Post-Secondary'},'Fedu':{0:'None', 1:'Primary', 2:'Grade 5-9', 3:'Secondary', 4:'Post-Secondary'}, 'subject':{'M':'Math', 'P':'Portuguese'}, 'studytime':{0:'<2 hr', 1:'2-5 hr', 2:'5-10 hr', 3:'>10 hr'}, 'school':{'GP':'Gabriel Peireira', 'MS':'Mousinho da Silveira'}}
cleanData = cleanData.replace(value_names)
col_names = {'paid':'extra_classes', 'Medu':'mother\'s_education', 'Fedu':'father\'s_education', 'famsup':'family_support', 'studytime':'study_time'}
cleanData = cleanData.rename(columns=col_names)
```

## Data Analysis (Task 4)

After loading and examining the data certain trends become very apparent. First I'll start by examining the grade distribution per study time category

```python
fig, ax = plt.subplots(1,3, figsize=(20,5))
sns.violinplot(x='study_time',y='G1',data=cleanData, hue='subject',split=True, order=['<2 hr','2-5 hr','5-10 hr','>10 hr'], ax=ax[0])
sns.violinplot(x='study_time',y='G2',data=cleanData, hue='subject',split=True, order=['<2 hr','2-5 hr','5-10 hr','>10 hr'], ax=ax[1])
sns.violinplot(x='study_time',y='G3',data=cleanData, hue='subject',split=True, order=['<2 hr','2-5 hr','5-10 hr','>10 hr'], ax=ax[2])
fig.suptitle('Grade Distribution Across School Periods by Weekly Study Time')
fig.show()
```

From the data above we can see that studytime does have a moderate effect on student performance in terms of their grades per period. For each later quarter there seem to be more observations in the lower portion of the grade axis for all of the study time categories. However, upper tail grade observations seem to be more prevalent in the higher study time categories, especially in the later quarters. What is interesting to note however is that the interquartile range for study times that are <2hr are ssignificantly lower than the other categories. Additionally, median and IQR scores seem to be similar for the 5-10 hr and >10 hr categories, indicating some diminishing returns to study time. Another observation of significance is that there appears to be less occurences of lower scores in the higher study categories. This could indicate that students that study more often, do indeed perform better. 

Now that we have established that students do indeed perform better with more study time, another pertinent question is the relation to study time and the education levels of the student's parents. 

```python
fig, ax = plt.subplots(1,2, figsize=(20,5))
ax[0].tick_params(axis='x', rotation = 25,)
ax[1].tick_params(axis='x', rotation = 25,)
sns.countplot(x='study_x_mother\'s_education', data = cleanData, hue ='subject', order=['<2 hr X None','2-5 hr X None', '5-10 hr X None', '>10 hr X None', '<2 hr X Primary','2-5 hr X Primary', '5-10 hr X Primary', '>10 hr X Primary', '<2 hr X Grade 5-9','2-5 hr X Grade 5-9', '5-10 hr X Grade 5-9', '>10 hr X Grade 5-9', '<2 hr X Secondary','2-5 hr X Secondary', '5-10 hr X Secondary', '>10 hr X Secondary', '<2 hr X Post-Secondary','2-5 hr X Post-Secondary', '5-10 hr X Post-Secondary', '>10 hr X Post-Secondary'], ax=ax[0]).set(xlabel="Study Time and Mother's Education Level", ylabel="Count")
sns.countplot(x='study_x_father\'s_education', data = cleanData, hue ='subject', order=['<2 hr X None','2-5 hr X None', '5-10 hr X None', '>10 hr X None', '<2 hr X Primary','2-5 hr X Primary', '5-10 hr X Primary', '>10 hr X Primary', '<2 hr X Grade 5-9','2-5 hr X Grade 5-9', '5-10 hr X Grade 5-9', '>10 hr X Grade 5-9', '<2 hr X Secondary','2-5 hr X Secondary', '5-10 hr X Secondary', '>10 hr X Secondary', '<2 hr X Post-Secondary','2-5 hr X Post-Secondary', '5-10 hr X Post-Secondary', '>10 hr X Post-Secondary'], ax=ax[1]).set(xlabel="Study Time and Father's Education Level", ylabel="Count")
fig.suptitle('Count of Study Time X Parent\'s Education Level')
fig.show()
```

A countplot of the cartesian product of shows the number of occurrences for each parents education level and the study habits of each student. There seems to be no real relation between the study time and parents education level as the graph is fairly evenly distributed. One note of importance is that the parents with post secondary education tended to have couts that were lower for higher study times which meant that the students study habits may have much more to do with their own motivation than their parents academic achievement.

The last factor I wish to analyze is the correlation between student's final grades and their educational support at home as well as their aspirations toward future education

```python
fig, ax = plt.subplots(1,2, figsize=(20,5))
sns.boxenplot(y='G3', x='higher', data=cleanData, hue='subject', ax=ax[0]).set(xlabel="Higher Education Aspirations")
sns.boxenplot(y='G3', x='family_support', data=cleanData, hue='subject', ax=ax[1]).set(xlabel="Family Support")
fig.suptitle('Aspirations Toward Higher Education and Final Grade as well as Family Support and Final Grade')
fig.show()
```

The family support graph provided a fairly even distribution as most of the grades were evenly interspersed between yes and no answers. Overall grades seemed to be slightly higher in the yes category with more outliers in the portuguese class. Math grades were fairly consistent with both categories. Overall Iconcluded that family support had no real effect for math grades, however portuguese grades seemed to benefit slightly. The higher education graph yielded far more distinct results. Overall grades in both subjects were drastically lower, especially in math. The yes distribution had a much higher downward skew with most of the observations appearing in the upward tail of the plot. This means that this category is a huge indicator for student's performance.

Overall it is reasonable to conclude that student's home relationship is a large factor in their scholarly success. Study time as well as aspirations towards future education seemed to be the main key factors toward student success. However, it is worth mentioning that much of the data used is categorical in nature.
